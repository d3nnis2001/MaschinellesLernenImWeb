## Exercise No. 3 - Clustering

```python
# Generating blobs with sklearn and plotting them

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AffinityPropagation, KMeans, SpectralClustering
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import make_blobs
from sklearn.metrics import silhouette_score

def gen_data(n_cluster):
    X, y = make_blobs(n_samples=1000, centers=n_cluster, n_features=2, random_state=61)
    return X, y

X, y = gen_data(10)
plt.scatter(X[:,0], X[:,1], s=1, c=y, cmap='viridis')
```

```python
# Clustering models and plotting them

models = [
    KMeans(n_clusters=10, init='k-means++', random_state=69), 
    AffinityPropagation(random_state=69), 
    SpectralClustering(n_clusters=10, random_state=69)
]

for model in models: 
    pred = model.fit_predict(X)
    plt.scatter(X[:,0], X[:,1], s=1, c=pred, cmap='viridis')
    plt.title(model.__class__.__name__)
    plt.show()
    print(f'Silhouette score: {silhouette_score(X, pred)}')
```
